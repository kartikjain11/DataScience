```python
import pandas as pd
import yfinance as yf
from rich import print as rprint
```

```python
apple = yf.Ticker("AAPL")
rprint(apple)
```

```python
apple_info = apple.info
rprint(apple_info)
```

```python
country = apple_info['country']
rprint(country)
```

```python
apple_share_price_data = apple.history(period='max')
apple_share_price_data.reset_index(inplace=True)
rprint(type(apple_share_price_data))
rprint(apple_share_price_data)
```

```python
apple_share_price_data.plot(x='Date',y='Open')
```

```python
dividends = apple.dividends
rprint(type(dividends))
rprint(dividends)
```

```python
dividends.plot()
```

```python
amd = yf.Ticker('AMD')
amd_info = amd.info
rprint(amd_info)
```

```python
rprint(amd_info['country'])
rprint(amd_info['sector'])
```

```python
data = amd.history(period='max')
data.reset_index(inplace=True)
rprint(data['Volume'][0])
rprint(data)
data.plot(x='Date',y='Open')
```
